```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("../data/Fish.csv")
df.head()
```

```python
species_map = {
    "Bream": "Tilapia",
    "Roach": "Tilapia",
    "Perch": "Nile Perch",
    "Pike": "Nile Perch",
    "Smelt": "Trout",
    "Parkki": "Catfish",
    "Whitefish": "Trout"
}

df["Species"] = df["Species"].map(species_map)
```

```python
df
```

```python
import numpy as np

def generate_species_environment():
    return {
        "Tilapia": {
            "temperature": np.random.normal(27, 2),
            "oxygen": np.random.normal(6, 1),
            "ph": np.random.uniform(6.5, 9),
            "density": np.random.randint(10, 40),
            "feeding_rate": np.random.uniform(2, 5)
        },
        "Nile Perch": {
            "temperature": np.random.normal(28, 1.5),
            "oxygen": np.random.normal(6, 0.8),
            "ph": np.random.uniform(6.5, 8.5),
            "density": np.random.randint(5, 20),
            "feeding_rate": np.random.uniform(3, 6)
        },
        "Catfish": {
            "temperature": np.random.normal(29, 2),
            "oxygen": np.random.normal(4, 1),
            "ph": np.random.uniform(6, 8),
            "density": np.random.randint(20, 60),
            "feeding_rate": np.random.uniform(3, 8)
        },
        "Trout": {
            "temperature": np.random.normal(15, 3),
            "oxygen": np.random.normal(9, 1),
            "ph": np.random.uniform(6.5, 8),
            "density": np.random.randint(5, 25),
            "feeding_rate": np.random.uniform(1, 4)
        }
    }
```

```python
species_env = generate_species_environment()

def assign_env(row):
    return pd.Series(species_env[row["Species"]])
```

```python
env_df = df.apply(assign_env, axis=1)

df = pd.concat([df, env_df], axis=1)
```

```python
df
```

```python
cols_to_remove = ["Weight", "Height", "Width", "Length1", "Length2", "Length3"]

df = df.drop(columns=[col for col in cols_to_remove if col in df.columns])
```

```python
df
```

```python
def generate_initial_weight(species):
    if species == "Tilapia":
        return np.random.uniform(5, 100)
    elif species == "Catfish":
        return np.random.uniform(10, 200)
    elif species == "Trout":
        return np.random.uniform(2, 80)
    elif species == "Nile Perch":
        return np.random.uniform(10, 150)
```

```python
df["initial_weight"] = df["Species"].apply(generate_initial_weight)

df
```

```python
df["days"] = np.random.randint(7, 120, size=len(df))
df
```

```python
def compute_growth(row):
    # Base growth rate per species
    if row["Species"] == "Tilapia":
        base_rate = 0.015
    elif row["Species"] == "Catfish":
        base_rate = 0.02
    elif row["Species"] == "Trout":
        base_rate = 0.012
    elif row["Species"] == "Nile Perch":
        base_rate = 0.018

    # Environmental influence
    temp_factor = row["temperature"] / 28
    oxygen_factor = row["oxygen"] / 6

    growth_factor = temp_factor * oxygen_factor

    # Final growth
    return row["initial_weight"] * (
        1 + base_rate * row["days"] * growth_factor
    )
```

```python
df["target_weight"] = df.apply(compute_growth, axis=1)
df
```

```python
df["target_weight"] += np.random.normal(0, 5, len(df))
df
```

```python
df["target_weight"] = df["target_weight"].clip(lower=df["initial_weight"])
df
```

```python
df.to_csv("../data/fish_growth_prediction.csv", index=False)
```
